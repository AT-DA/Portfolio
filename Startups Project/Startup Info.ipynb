```python
#Now i'll perform analysis on the category and category group columns then export them separately for visualization.
```

```python
#Analysing the category group columns
```

```python
cg = df['Category Groups'].str.split(',', expand = True)
```

```python
cg.stack().to_frame(name = 'Category Groups').reset_index(level = 1, drop = True).apply(lambda x: x.str.strip()).join(df[['total funding amount', 'Number of Investors']]).groupby(['Category Groups']).sum().sort_values('total funding amount', ascending = False)
#.stack()- stack all Category Groups columns into one column
#.to_frame(name = 'Caregory Groups')- convert the column to a dataframe and name the column 'Caregory Groups'
#.reset_index(level = 1, drop = True)- fix the index so that it can be concatenated with other dataframes based on the index of rach row.
#.apply(lambda x: x.str.strip())- remove leading and trailing whitespace of all columns because whitespace might lead to duplication
#.join(df[['total funding amount']])- add the total funding amount beside each category group. The joining is based on the index number.
#.groupby(['Caregory Groups']).sum().sort_values('total funding amount', ascending = False)- Group by the category groups column created in the first two steps then sort descending
```

```python
cg = cg.stack().to_frame(name = 'Category Groups').reset_index(level = 1, drop = True).apply(lambda x: x.str.strip()).join(df[['total funding amount', 'Number of Investors']]).groupby(['Category Groups']).sum().sort_values('total funding amount', ascending = False)
```

```python
cg.to_excel(r'C:\Users\Abood\Desktop\Data Analysis Practice\Python Data Analysis Practice\Startups Project\Category Groups.xlsx')
#Export the results in a separate file for visualization.
```
