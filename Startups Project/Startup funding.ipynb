```python
df['Last Funding Amount - ORIG'].value_counts()
```

```python
df['Last Funding Amount - ORIG'] = df['Last Funding Amount - ORIG'].astype(str)  #First make sure all rows in the column 'Last Funding Amount - ORIG' are set as string, or else if i use the str.replace function before conversion, it will return nan
```

```python
df['Last Funding Amount - ORIG'] = df['Last Funding Amount - ORIG'].str.replace(',','')  
```

```python
c = df['Last Funding Amount - ORIG'].str.extract('(\D+)', expand = False).unique().tolist()
```

```python
''.join(c[1:])  #Concatenating a list. I used c[1:] because if i start with the first element which is 'nan', it will return an error.
```

```python
df['Last Funding Amount - ORIG'].str.split('(\D+)', expand = True)
```

```python
df['m'] = df['Last Funding Amount - ORIG'].str.split('(\D+)', expand = True)[0] #column containing the numbers without the currencies
```

```python
df['n'] =  df['Last Funding Amount - ORIG'].str.split('(\D+)', expand = True)[2] #column containing the numbers after they have been stripped of their currencies
```

```python
df.info()
```

```python
df['n'].fillna('', inplace = True) #This is done to fill the rows containing 'None' with blank. If i dont do this step, the result will be Nan when adding columns m and n because a number cant be added with the word None.
```

```python
df['last funding amount'] = df['m'] + df['n']
```

```python
df.drop(columns = ['m','n'], inplace = True)
```

```python
df.info()
```
