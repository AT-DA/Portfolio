```python
#1. What are the genres that have the highest average rating?
```

```python
g = df['genres'].str.split('|', expand = True)
```

```python
g.stack().to_frame(name = 'genres').reset_index(level = 1, drop = True).join(df[['title','vote_average']]).groupby(['genres']).mean().sort_values('vote_average', ascending = False).head()
#.stack()- stack all genre columns into one column
#.to_frame(name = 'genres')- convert the column to a dataframe and name the column 'genres'
#.reset_index(level = 1, drop = True)- fix the index so that it can be concatenated with other dataframes based on the index of rach row.
#.join(df[['title','vote_average']])- add the title and the vote_average beside each genre. The joining is based on the index number.
#.groupby(['genres']).mean().sort_values('vote_average', ascending = False)- Group by the genres column created in the first two steps then sort descending
```

```python
#2. What are the plot keywords that have the highest average rating?
```

```python
k = df['keywords'].str.split('|', expand = True)
```

```python
k.stack().to_frame(name = 'keywords').reset_index(level = 1, drop = True).join(df[['title','vote_average']]).groupby(['keywords']).mean().sort_values('vote_average', ascending = False).head()
```

```python
#3. Who are the highly rated directors?
```

```python
df[['title', 'director','vote_average']].groupby('director').mean().sort_values(['vote_average'], ascending = False).head()
```

```python
#4. Who are the highly rated actors?
```

```python
a = df[['actor_1', 'actor_2', 'actor_3']]
```

```python
a.stack().to_frame(name = 'actors').reset_index(level = 1, drop = True).join(df[['title','vote_average']]).groupby(['actors']).mean().sort_values('vote_average', ascending = False).head()
```

```python
#5. Who are the most profitable directors?
```

```python
df[['title', 'director','profit']].groupby('director').sum().sort_values('profit', ascending = False).head()
```

```python
#6. Who are the most profitable actors?
```

```python
a = df[['actor_1', 'actor_2', 'actor_3']]
```

```python
a.stack().to_frame(name = 'actors').reset_index(level = 1, drop = True).join(df[['title','profit']]).groupby(['actors']).sum().sort_values('profit', ascending = False).head()
```

```python
#7. Which countries have higher average rating?
```

```python
c = df['production_countries'].str.split('|',expand = True)
```

```python
c.stack().to_frame(name = 'countries').reset_index(level = 1, drop = True).join(df[['title','vote_average']]).groupby(['countries']).mean().sort_values('vote_average', ascending = False).head()
```

```python
#8. Which countries produced more movies?
```

```python
c = df['production_countries'].str.split('|',expand = True)
```

```python
c.stack().to_frame(name = 'countries').reset_index(level = 1, drop = True).join(df['title']).groupby(['countries']).count().sort_values('title', ascending = False).head()
```
